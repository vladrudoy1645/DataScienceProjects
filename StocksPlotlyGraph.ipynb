```python
import pandas as pd
from pandas_datareader import data,wb
import numpy as np
import datetime

import matplotlib.pyplot as plt
% matplotlib inline

plt.rcParams['axes.titlesize'] = 'x-large'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams.update({'font.size' : 15})
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = data.DataReader("BAC", 'yahoo', start, end)['Close']
GS  = data.DataReader("GS", 'yahoo', start, end)['Close']
C = data.DataReader("C", 'yahoo', start, end)['Close']
MS = data.DataReader("MS", 'yahoo', start, end)['Close']
WFC = data.DataReader("WFC", 'yahoo', start, end)['Close']
JPM = data.DataReader("JPM", 'yahoo', start, end)['Close']

tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
days = pd.date_range(start, end, freq='D')
```

```python
import plotly.plotly as py
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
def plotfunc(banklist, colorlist,symbols):
    dat = []
    for bank,color,ticker in zip(banklist, colorlist,tickers):
        trace = go.Scatter(
            x = days,
            y = bank,
            name = '{} Stock Prices'.format(ticker),
                line = dict(
                    color = (color),
                    width = 2))
        dat.append(trace)
    return iplot({"data": dat,
                     "layout": go.Layout(title="Closing Stock Prices for Six Major US Banks",
                                        xaxis = dict(title = 'Date'),
                                          yaxis = dict(title = 'Stock Price ($)'))})       
```

```python
banks = [BAC, C, GS, JPM, MS, WFC]
colorlist = ['rgb(220,20,60)','rgb(22, 96, 167)', 'rgb(135,206,250)','rgb(0,0,128)', 'rgb(119,136,153)', 'rgb(255,215,0)']
```

```python
plotfunc(banks,colorlist,tickers)
```
